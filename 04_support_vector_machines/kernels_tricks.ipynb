```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.svm import SVC
```

```python
def plot_decision_boundary(model, X, y, title):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    xx = np.linspace(xlim[0], xlim[1], 200)
    yy = np.linspace(ylim[0], ylim[1], 200)

    XX, YY = np.meshgrid(yy, xx)

    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = model.decision_function(xy).reshape(XX.shape)

    ax.contour(XX, YY, Z, levels=[0], linewidths=2, colors='k')
    plt.title(title)
    plt.show()
    
```

```python
X, y = make_moons(n_samples=100, noise=0.2, random_state=42)

# Train with different kernels
kernels = ['linear', 'poly', 'rbf']
for kernel in kernels:
    model = SVC(kernel=kernel, degree=3 if kernel == 'gamma' else 3, gamma='scale')
    model.fit(X, y)
    plot_decision_boundary(model, X, y, f'SVM with {kernel} kernel')
```
